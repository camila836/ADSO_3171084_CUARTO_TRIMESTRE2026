```python
class Libro:
  def __init__(self, titulo, precio):
    self.__titulo = titulo
    self.__precio = precio
    #decoradores para el getter, osea, mostrar los valores de las variables
    @property
    def titulo(self):
     return self.__titulo

    @property #decorador para el getter
    def precio(self):
      return self.__precio

    #setter, modificar los valores de los atributos

    @titulo.setter
    def titulo(self, nuevo_titulo):
      if isinstance(nuevo_titulo, str):
        self.__titulo = nuevo_titulo
      else:
        print("Debe ingresar un nuevo texto")

    @precio.setter
    def precio(self, nuevo_precio):
      if isinstance(nuevo_precio, (int, float)):
        self.__precio = nuevo_precio
      else:
        print("Debe ingresar un numero entero o flotante")

    def mostrar_info(self):
      print(f"El titulo del libro es {self.titulo}, el precio es: ${self.precio:,.2f}")

  def main():
    print("Programa para Verificar los Decoradores en Programación Orientada a Objetos @property para GETTER y @atributo.setter para SETTER")
    libro1 = Libro("100 años de soledad",75000)

    #Mostrar informacion
    libro1.mostrar_info()

    #Actualizar titulo
    libro1.titulo = "El coronel no tiene quien le escriba"
    print(libro1.titulo)

    #Actualizar precio correctamente
    libro1.precio = float(input("Ingrese el nuevo valor del libro $,,"))
    print(libro1.precio)
    libro1.mostrar_info()

  if __name__ == "__main__":
    main()
```
