```python
class Estudiante:

    def __init__(self, nombre, documento, edad, programa, promedio):
        self.__nombre = nombre
        self.__documento = documento
        self.__edad = edad
        self.__programa = programa
        self.__promedio = promedio

    # Métodos para mostrar atributos
    def mostrar_nombre(self):
        return self.__nombre

    def mostrar_documento(self):
        return self.__documento

    def mostrar_edad(self):
        return self.__edad

    def mostrar_programa(self):
        return self.__programa

    def mostrar_promedio(self):
        return self.__promedio

    # Métodos para modificar atributos
    def modificar_nombre(self, nuevo_nombre):
        self.__nombre = nuevo_nombre

    def modificar_documento(self, nuevo_documento):
        self.__documento = nuevo_documento

    def modificar_edad(self, nueva_edad):
        self.__edad = nueva_edad

    def modificar_programa(self, nuevo_programa):
        self.__programa = nuevo_programa

    def modificar_promedio(self, nuevo_promedio):
        self.__promedio = nuevo_promedio

    # Método para mostrar toda la información
    def mostrar_informacion(self):
        return f"""
        Nombre: {self.__nombre}
        Documento: {self.__documento}
        Edad: {self.__edad}
        Programa: {self.__programa}
        Promedio: {self.__promedio}
        """

def main():
  estudiante1 = Estudiante("Christi", "1014667088", "18", "ADSO", "4.5")
  print("Informacion inicial del estudiante:")
  print(estudiante1.mostrar_informacion())

  estudiante1.modificar_nombre("Noemi")
  estudiante1.modificar_programa("Moda")
  estudiante1.modificar_documento("45455454")
  print("Información actualizada del estudiante")
  print(estudiante1.mostrar_informacion())

# Llamar a la función main
main()
```
