```python
class Libro:
  def __init__(self, titulo, estatura):
    self.__titulo = titulo
    self.__estatura = estatura

  def get_estatura(self):
    return self.__estatura

  def set_estatura(self, nuevo_estatura):
    if isinstance(nuevo_estatura, (int, float)) and nuevo_estatura > 0:
      self.__estatura = nuevo_estatura
    else:
      print("El estatura debe ser un numero positivo")

  def mostrar_info(self):
    print(f"\nTitulo: {self.__titulo}")
    print(f"estatura: {self.__estatura:,.2f}")
    print("--------------------------------")
    print(f"Titulo: {self.__titulo}, estatura {self.__estatura:,.2f}\n")

def main():
  libro1 = Libro("Python Profesional", 120000)
  libro2 = Libro("Java Profesional", 150000)

  print("Informacion inicial del libro:")
  libro1.mostrar_info()

  print("estatura actual:", libro1.get_estatura())
  libro1.set_estatura(-10)

  libro1.set_estatura(120000)

  print("Nuevo estatura:", libro1.get_estatura())
  print("\nPrograma finalizado")

if __name__ == "__main__":
  main()
```
