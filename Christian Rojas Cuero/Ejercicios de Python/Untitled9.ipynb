```python
class Opera:
  def __init__(self, a, b):
    self.a = a
    self.b = b
    print("Inicializando Opera")
  def calculo(self):
    print("Calculando Opera")

class Potenciar(Opera):
  def __init__(self, a, b):
    super().__init__(a, b)
    print("Preparando opera potencia: ")
    if self.b != 3:
      resultados = self.a ** self.b
      print(f"La potencia es: {resultados}")
    else:
      print("No se puede potenciar por encima de 3.")

    return resultados


def main():
  a = float(input("Ingrese el primer dato: "))
  b = float(input("Ingrese el segundo dato: "))
  opera1 = Potenciar(a, b)
  opera1.calculo()

if __name__ == "__main__":
  main()
```
