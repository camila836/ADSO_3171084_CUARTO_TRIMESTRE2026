```python
class Libro:
    def __init__(self, titulo, precio):
        self.__titulo = titulo
        self.__precio = precio

    # ── GETTERS (decorador @property) ──────────────────────────────
    @property
    def titulo(self):
        return self.__titulo

    @property
    def precio(self):
        return self.__precio

    #  SETTERS
    @titulo.setter
    def titulo(self, nuevo_titulo):
        if isinstance(nuevo_titulo, str) and nuevo_titulo.strip():
            self.__titulo = nuevo_titulo
        else:
            print("El título debe ser un texto no vacío.")

    @precio.setter
    def precio(self, nuevo_precio):
        if isinstance(nuevo_precio, (int, float)) and nuevo_precio >= 0:
            self.__precio = nuevo_precio
        else:
            print("El precio debe ser un número entero o flotante mayor o igual a 0.")

    #MOSTRAR INFORMACIÓN
    def mostrar_info(self):
        print("─" * 40)
        print(f"  Título : {self.__titulo}")
        print(f"  Precio : ${self.__precio:,.2f}")
        print("─" * 40)

# REPRESENTACIÓN DEL OBJETO
    def __str__(self):
        return f"Libro('{self.__titulo}', ${self.__precio:,.2f})"

    def __repr__(self):
        return f"Libro(titulo='{self.__titulo}', precio={self.__precio})"


# MAIN
def main():
    libro1 = Libro("Python Profesional", 120000)
    libro2 = Libro("Java Profesional",   150000)

    print("=== Información inicial ===")
    libro1.mostrar_info()
    libro2.mostrar_info()

    # Usar getter con @property  →  sin paréntesis
    print("Título actual :", libro1.titulo)
    print("Precio actual :", libro1.precio)

    # Intentos NO válidos (disparan validación del setter)
    print("\n--- Intentos inválidos ---")
    libro1.precio = -500          # número negativo
    libro1.titulo = ""            # texto vacío
    libro1.titulo = 123           # no es string

    # Actualización CORRECTA con setter
    libro1.titulo = "Python Avanzado"
    libro1.precio = 135_000

    print("\n=== Información actualizada ===")
    libro1.mostrar_info()

    # Mostrar representación str y repr
    print("\n--- Representaciones del objeto ---")
    print(str(libro1))
    print(repr(libro1))

    print("\nPrograma finalizado.")


if __name__ == "__main__":
    main()
```

```python
# CLASE PADRE
class Vehiculo:

    def __init__(self, marca, color, velocidad):
        self.__marca     = marca
        self.__color     = color
        self.__velocidad = velocidad

    # GETTERS
    def get_marca(self):
        return self.__marca

    def get_color(self):
        return self.__color

    def get_velocidad(self):
        return self.__velocidad

    # SETTERS
    def set_marca(self, nueva_marca):
        if isinstance(nueva_marca, str) and nueva_marca.strip():
            self.__marca = nueva_marca
        else:
            print("La marca debe ser un texto no vacío.")

    def set_color(self, nuevo_color):
        if isinstance(nuevo_color, str) and nuevo_color.strip():
            self.__color = nuevo_color
        else:
            print("El color debe ser un texto no vacío.")

    def set_velocidad(self, nueva_velocidad):
        if isinstance(nueva_velocidad, (int, float)) and nueva_velocidad > 0:
            self.__velocidad = nueva_velocidad
        else:
            print("La velocidad debe ser un número mayor a 0.")

    def mover(self):
        print(f"El vehículo '{self.__marca}' se está moviendo a {self.__velocidad} km/h.")


# CLASE HIJA
class Moto(Vehiculo):

    def __init__(self, marca, color, velocidad, tipo):
        super().__init__(marca, color, velocidad)
        self.__tipo = tipo

    def get_tipo(self):
        return self.__tipo

    def set_tipo(self, nuevo_tipo):
        if isinstance(nuevo_tipo, str) and nuevo_tipo.strip():
            self.__tipo = nuevo_tipo
        else:
            print("El tipo debe ser un texto no vacío.")

    def mover(self):  # SOBREESCRIBIMOS EL MÉTODO DEL PADRE
        print(f"La moto '{self.get_marca()}' de tipo '{self.__tipo}' se mueve a {self.get_velocidad()} km/h.")


# MAIN
if __name__ == "__main__":

    # Crear un vehículo
    v = Vehiculo("Toyota", "Rojo", 120)
    v.mover()

    # Usar setters
    v.set_velocidad(150)
    v.mover()

    # Crear una moto
    m = Moto("Yamaha", "Negro", 180, "Deportiva")
    m.mover()

    # Modificar datos de la moto
    m.set_color("Azul")
    m.set_velocidad(200)
    m.mover()
```

```python
class Operacion:
    def __init__(self, dato1, dato2): # CONSTRUCTOR
      self.dato1 = dato1
      self.dato2 = dato2

class Suma(Operacion):
    def calcular(self):
      resultado= self.dato1 + self.dato2
      print(f"LA SUMA ES: {resultado}")


class Resta(Operacion):
    def calcular(self):
      resultado= self.dato1 - self.dato2
      print(f"LA RESTA ES {resultado}")


class Multiplicacion(Operacion):
    def calcular(self):
      resultado= self.dato1 * self.dato2
      print(f"LA MULTIPLICACION ES {resultado}")


class Division(Operacion):
    def calcular(self):
      if self.dato2 !=0:
        resultado = self.dato1 / self.dato2
        print(f"LA DIVISION ES{resultado}")
      else:
        print("EL SEGUNDO VALOR O NUMERO ES INVALIDO")


def main():
  dato1 = int(input("INGRESE EL PRIMER NUMERO"))
  dato2 = int(input("INGRSE EL SEGUNDO NUMERO"))
  operacion1=Suma(dato1, dato2)
  operacion1.calcular()
  operacion2=Resta(dato1, dato2)
  operacion2.calcular()
  operacion3=Multiplicacion(dato1, dato2)
  operacion3.calcular()
  operacion4=Division(dato1, dato2)
  operacion4.calcular()

if __name__ == "__main__":
    main()
```
